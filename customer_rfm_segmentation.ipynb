```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Set display options
pd.set_option('display.max_columns', None)
```

```python
# Load dataset
df = pd.read_csv("data.csv", encoding='cp1252')
df.head()
```

```python
# Remove rows with missing CustomerID
df = df[df['CustomerID'].notnull()]

# Remove cancelled orders (Quantity <= 0)
df = df[df['Quantity'] > 0]

# Convert InvoiceDate to datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Calculate TotalPrice
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']

# Check result
df.head()
```

```python
# Set snapshot date (analysis reference point)
snapshot_date = df['InvoiceDate'].max() + pd.Timedelta(days=1)
print(f"Snapshot date: {snapshot_date.date()}")
```

```python
# RFM calculation
rfm = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (snapshot_date - x.max()).days,  # Recency
    'InvoiceNo': 'nunique',                                   # Frequency
    'TotalPrice': 'sum'                                       # Monetary
})

# Rename columns
rfm.rename(columns={
    'InvoiceDate': 'Recency',
    'InvoiceNo': 'Frequency',
    'TotalPrice': 'Monetary'
}, inplace=True)

# View result
rfm.head()
```

```python
# RFM score for each metric using quartiles
rfm['R'] = pd.qcut(rfm['Recency'], 4, labels=[4, 3, 2, 1]).astype(int)
rfm['F'] = pd.qcut(rfm['Frequency'].rank(method='first'), 4, labels=[1, 2, 3, 4]).astype(int)
rfm['M'] = pd.qcut(rfm['Monetary'], 4, labels=[1, 2, 3, 4]).astype(int)

# Create RFM segment and score
rfm['RFM_Segment'] = rfm['R'].astype(str) + rfm['F'].astype(str) + rfm['M'].astype(str)
rfm['RFM_Score'] = rfm[['R', 'F', 'M']].sum(axis=1)

rfm.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
rfm['RFM_Score'].value_counts().sort_index().plot(kind='bar')
plt.title('Customer Distribution by RFM Score')
plt.xlabel('RFM Score')
plt.ylabel('Number of Customers')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Best customers
best_customers = rfm[rfm['RFM_Score'] >= 10]

# At risk
at_risk = rfm[(rfm['R'] == 1) & (rfm['F'] >= 2)]

# Loyal customers (часто покупают)
loyal_customers = rfm[rfm['F'] == 4]

# Print group sizes
print(f"Best customers: {len(best_customers)}")
print(f"At risk customers: {len(at_risk)}")
print(f"Loyal customers: {len(loyal_customers)}")
```

```python
# Pivot table for heatmap
rfm_pivot = rfm.pivot_table(index='F', columns='R', values='Monetary', aggfunc='mean')

# Heatmap
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.heatmap(rfm_pivot, annot=True, fmt=".1f", cmap="YlGnBu")
plt.title("Heatmap of Monetary Value by R and F Scores")
plt.xlabel("Recency Score")
plt.ylabel("Frequency Score")
plt.tight_layout()
plt.show()
```
